```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.font_manager import FontProperties
from DataMonoClass import DataMono
from CFST_Cir_Section_Analysis import CFST_Cir_Section_Analysis

def cutData(data, ratio=0.4):
    DD, N, M, Nc, Mc, Ns, Ms = data
    threhold = 0.4 * M.max()
    for i in range(M.size):
        if M[i] >= threhold:
            break
    return data[:,i:]

def remove_nan(data1, data2):
    out = np.zeros((1,2))
    for i in range(data1.size):
        if not (np.isnan(data1[i]) or np.isnan(data2[i])):
            out = np.vstack((out, [data1[i], data2[i]]))
    out = out[1:].T
    return out[0], out[1]

def countUp(k, b, datax, datay):
    f = np.poly1d([k ,b])
    err = datay - f(datax)
    a = err[err>0.0]
    return len(a) / datax.size * 100
```

```python
cutRatio = 0.5
iniRatio = 0.4
targetDuctility = 5
loadLevel = [0, 0.05, 0.1, 0.2, 0.3]

conditions = np.loadtxt('_CFST_Conditions.txt')
path = './DataC'


for i in range(len(loadLevel)):
    curLevel = loadLevel[i]
    dataAll = np.load('{0}/level-{1:.0f}/MaterialRelation.npy'.format(path, curLevel*100))
    outputs = np.zeros((1,9))
    for j in range(len(dataAll)):
        D, t, L, fc, fy = conditions[j]
        curData = dataAll[j]
        DD, fccs, fybs, hcs, hss = curData.T

        fcData = DataMono(DD, fccs, iniRatio)
        fyData = DataMono(DD, fybs, iniRatio)
        temp, fcc = fcData.findYield1()
        temp, fyb = fyData.findYieldByDuctility(targetDuctility)
        fcc = min(fcc, fccs[-1])
        outputs = np.vstack((outputs, [D, t, L, fc, fy, fcc, fyb, fcc/fc, fyb/fy]))
    outputs = outputs[1:]
    np.savetxt('{0}/level-{1:.0f}/fc_fy.txt'.format(path, curLevel*100), outputs)
```

```python


font_label = FontProperties(family='Times New Roman',style='normal', size = 10)
font = {'family' : 'Times New Roman',
        'weight' : 'normal',
        'size'   : 10,
        }
matplotlib.rc('font', **font)

loadLevel = [0, 0.05, 0.1, 0.2, 0.3]
path = './DataC'
err_s = np.array([0.036,0.032,0.026,0.015,0.015])
err_c = 0.126
point_styles = ['k^', 'r>', 'bo', 'yD', 'gs']
colors = [item + '-' for item in point_styles]
ms = 5
x_range = 1, 4.5
y_range = 0.9, 1.2
fig_c, ax_c = plt.subplots(1, 1, figsize=[6.2,4])
fig_s, ax_s = plt.subplots(1, 1, figsize=[6.2,4])
factorS = np.zeros((1,2))

for i in range(len(loadLevel)):
    curLevel = loadLevel[i]
    data = np.loadtxt('{0}/level-{1:.0f}/fc_fy.txt'.format(path, curLevel*100)).T
    D, t, L, fc, fy, fcc, fyb, ratioFc, ratioFy = data
    #factor
    A = np.pi * D**2 / 4.0
    Ac = np.pi * (D - 2.0 * t)**2 / 4.0
    As = A - Ac
    factor = np.log(Ac * fc / As /fy**0.5)
    #plot
    ax_c.plot(factor, ratioFc, point_styles[i], markersize=ms)
    factorC = np.append(factorC, factor) if i!=0 else factor
    ratioC = np.append(ratioC, ratioFc) if i!=0 else ratioFc


    ax_s.plot(factor, ratioFy, point_styles[i], markersize=ms)
    ps, Vs = np.polyfit(factor.ravel(),ratioFy.ravel(),1, cov=True)
    ks, bs = ps
    bs -= err_s[i]
    print(countUp(ks, bs, factor, ratioFy))
    factorS = np.vstack((factorS, [ks, bs]))
    fs = np.poly1d([ks ,bs])
    ax_s.plot([-1,10], fs([-1,10]), colors[i], label =r'$\mu_n$={0:.0f}\%'.format(curLevel*100), markersize=ms, linewidth=3)

factorC, ratioC = remove_nan(factorC, ratioC)
pc, Vc = np.polyfit(factorC.ravel(),ratioC.ravel(),1, cov=True)
kc, bc = pc
bc -= err_c
print(countUp(kc, bc, factor, ratioFc))
fc = np.poly1d([kc ,bc])
ax_c.plot([-1,10], fc([-1,10]), 'k-', label ='Regression Line', markersize=ms, linewidth=5)
np.savetxt('factorC.out', [kc, bc])
factorS = factorS[1:]
np.savetxt('factorS.out', factorS)



ax_c.set_ylim(0.8,2.5)
ax_s.set_ylim(0.9,1.25)
for ax in [ax_c, ax_s]:
    ax.set_xlim(x_range)
    #ax.set_ylim(y_range)
    ax.grid(True)
plt.show()
print(factorS)
```

```python
bavg = np.mean(factorS[:,1])
levels = np.array([0.0, 0.05, 0.1, 0.2, 0.3])
a1, b1 = np.polyfit(levels, factorS[:,0], deg=1)
f1 = np.poly1d([a1 ,b1])
fig1, ax1 = plt.subplots(1,1)
ax1.plot(levels, factorS[:,0])
ax1.plot(levels, f1(levels))
np.savetxt('factorS1.out', [a1, b1, bavg])
```
